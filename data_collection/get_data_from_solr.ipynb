```python
import requests
def api_helper(num_row, start_row):
    cs_url  = "http://54.237.219.50:8984/solr/faers/select?indent=on&q=occr_country:US&rows=%s&start=%s&wt=csv" % (num_row, start_row)
    #cs_url  = 'http://54.237.219.50:8984/solr/faers/select?indent=on&q=occr_country:US&rows=30&start=20&wt=csv'
    print cs_url
    r = requests.get(cs_url)
    if r.status_code == 200:
        text_file = open("solr_us_all.csv", "a")
        text_file.write(r.content)
        text_file.close()
```

```python
def getDataFromSolr():
    cur_start = 0
    max_row = 3013736
    while cur_start < max_row:
        print 'cur_start =',cur_start
        api_helper(10000,cur_start)
        cur_start += 10000
    num_row = max_row - cur_start + 10000
    print 'last row num = ',num_row
    api_helper(num_row, cur_start-10000)
    print 'Successfully collected all the data'
```

```python
def main():
    getDataFromSolr()
    #getFirst10kFromSolr()
    return


if __name__ == '__main__':
    main()
```
