```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
wt = range(1,350)
```

```python
k = [10 ** i for i in np.arange(-2,1, 0.5)]
```

```python
x0 = 170
log_scale = pd.DataFrame()
for kv in k:
    print kv
    log_scale[kv] = [1/(1+np.e ** (-kv*(w-x0))) for w in wt]
    log_scale[kv].plot(legend = True)
plt.show()
```

```python
age = range(1,120)
x0 = 50
log_scale = pd.DataFrame()
for kv in k:
    print kv
    log_scale[kv] = [1/(1+np.e ** (-kv*(a-x0))) for a in age]
    log_scale[kv].plot(legend = True)
plt.show()
```
